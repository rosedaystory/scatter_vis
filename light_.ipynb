```python
def penetration_r (a_right, rate):
    b = np.zeros_like(a_right)
    a_right = a_right * (1 - rate)
    b[1:] = a_right[0:-1]
    b[0] = 100
    return b

def penetration_l (a_left, rate):
    b = np.zeros_like(a_left)
    a_left = a_left * (1 - rate)
    b[0:-1] = a_left[1:]
    b[-1] = 0
    return b
    
def to_right(a_right, a_left, rate):
    a = a_right + a_left
    b = np.zeros_like(a)
    a = a * 0.5 * rate
    b[1:] = a[0:-1]
    return b

def to_left(a_right, a_left, rate):
    a = a_right + a_left
    b = np.zeros_like(a)
    a = a * 0.5 * rate
    b[0:-1] = a[1:]
    return b

def tot(a, rate):
    b = a.copy()
    a[0] = penetration_r(b[0],rate) + to_right(b[0], b[1], rate)
    a[1] = penetration_l(b[1],rate) + to_left(b[0], b[1], rate)
    return a
```

```python
a = np.zeros(100)


def gen_res(rate, row, n):

    a = np.zeros((2,row))
    c = np.zeros((n,2,row))
    for i in range(n):
        a = tot(a, rate)
        c[i] = a
    return c
```
